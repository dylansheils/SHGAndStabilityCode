```python
import pandas as pd
```

```python
dfStiff = pd.read_csv('wren_s-0_t-1Stiffness.csv')
```

```python
dfThermo = pd.read_csv('wren_s-0_t-1Thermo.csv')
```

```python
dfPhonon = pd.read_csv('wren_s-0_t-1Phonons.csv')
```

```python
print(dfStiff)
```

```python
df = pd.concat([dfThermo, dfStiff, dfPhonon], axis=1)
```

```python
print(df)
```

```python
import numpy as np
```

```python
df["thermoAvg"] = (1/3)*(df["thermo_pred_n0"] + df["thermo_pred_n1"] + df["thermo_pred_n2"])
```

```python
df["stiffAvg"] = (1/3)*((np.exp(df["stiffness_pre-logits_c1_n0"])/(1 + np.exp(df["stiffness_pre-logits_c1_n0"]))) + (np.exp(df["stiffness_pre-logits_c1_n1"])/(1 + np.exp(df["stiffness_pre-logits_c1_n1"]))) + (np.exp(df["stiffness_pre-logits_c1_n2"])/(1 + np.exp(df["stiffness_pre-logits_c1_n2"]))))
```

```python
print(df)
```

```python
df["phononAvg"] = np.exp(df["stiffness_pre-logits_c1_n0"])/(1 + np.exp(df["stiffness_pre-logits_c1_n0"]))
```

```python
print(df)
```

```python
df = df.sort_values(by=["stiffAvg","phononAvg", "thermoAvg"], ascending=[False,False,True])
```

```python
print()
```

```python
df = df[~(df['thermoAvg'] > 0.06)]
```

```python
df = df[~(df['phononAvg'] < 0.75)]
```

```python
df = df[~(df['stiffAvg'] < 0.75)]
```

```python
print(df)
```

```python
df.to_csv("5kMBTFamilyPredictionsAll.csv")
```
