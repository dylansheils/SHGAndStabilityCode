```python
# Using readlines()
file1 = open('finalOutputForThermo.txt', 'r')
Lines = file1.readlines()
 
count = 0
# Strips the newline character
valid = []
train = []
for line in Lines:
    if("thermo" in line):
        if("Train" in line):
            train.append(float(line[line.find("-")+15:line.find("-")+20]))
        if("Validation" in line and "Baseline" not in line):
            valid.append(float(line[line.find("-")+15:line.find("-")+20]))
```

```python
import matplotlib.pyplot as plt
xpoints = []
for x in range(1, len(valid)+1):
    xpoints.append(x)
```

```python
plt.plot(xpoints, train, 'b', label="Training")
plt.plot(xpoints, valid, 'r', label="Validation")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(['Training', 'Validation'])
plt.show()
```
